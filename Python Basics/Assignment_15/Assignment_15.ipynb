# Q7

```python
def genPrimes():
    primes = [2]  # Initialize the list of prime numbers with 2
    yield 2

    num = 3
    while True:
        is_prime = True
        for prime in primes:
            if prime * prime > num:
                break
            if num % prime == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
            yield num
        num += 2  # Check only odd numbers as even numbers greater than 2 are not prime

# Example usage of genPrimes:
prime_generator = genPrimes()
for i in range(10):
    print(next(prime_generator))
```
