# Q1

RGBA stands for Red, Green, Blue, and Alpha. It is a color model used in digital image processing and computer graphics. In addition to the primary color channels (red, green, blue), the RGBA model includes an alpha channel that represents the level of transparency or opacity of a pixel.

# Q2

To get the RGBA value of an image using the Pillow module (Python Imaging Library, also known as PIL), you can use the getpixel() method. Here's an example:

```python
from PIL import Image
import matplotlib.pyplot as plt

# Open an image file
img = Image.open(r"D:\iNeuron\Python Basics\Assignment_14\Chess_Board.png")

# Display the image
plt.imshow(img, cmap = "gray")

# Get the RGBA value at a specific pixel (x, y)
rgba_value = img.getpixel((100, 100))

# 'rgba_value' will be a tuple like (R, G, B, A), where R, G, B, and A are integer values between 0 and 255.
print(f"The RGBA value of the image pixel at 100, 100 is {rgba_value}" )
```

# Q3

A box tuple in Pillow represents a rectangular region within an image. It is a tuple of four values: (left, upper, right, lower), where:
- left is the x-coordinate of the left edge of the box.
- upper is the y-coordinate of the upper edge of the box.
- right is the x-coordinate of the right edge of the box.
- lower is the y-coordinate of the lower edge of the box.

The box tuple defines a rectangular region within the image, where (left, upper) represents the top-left corner, and (right, lower) represents the bottom-right corner of the rectangle.

# Q5

To get an Image object for a 100×100 image, excluding the lower-left quarter of it, you can use the .crop() method:

```python
from PIL import Image
import matplotlib.pyplot as plt

# Open an image file
img = Image.open(r"D:\iNeuron\Python Basics\Assignment_14\Chess_Board.png")

# Crop the image to a 100x100 region starting from (0, 0)
cropped_img = img.crop((0, 0, 100, 100))

# Display the image
plt.imshow(cropped_img, cmap = "gray")
```

# Q6

After making changes to an Image object in Pillow, you can save it as an image file using the .save() method. For example, to save the modified image as a PNG file:

```python
cropped_img.save('cropped_image.png')
```

You can specify the file format by providing a file extension (e.g., .png, .jpg, .bmp, etc.) in the filename.

# Q8

Image objects in Pillow do not have built-in drawing methods. To perform drawing operations (e.g., lines, circles, rectangles) on an image, you need to create a drawing object from the ImageDraw module. Here's how you can create a drawing object for an image:

```python
from PIL import Image, ImageDraw

# Open an image file
img = Image.open(r"D:\iNeuron\Python Basics\Assignment_14\Chess_Board.png")

# Create a drawing object for the image
draw = ImageDraw.Draw(img)

# Now, you can use 'draw' to perform various drawing operations on 'img'.
```

The draw object allows you to draw shapes and text on the image.
