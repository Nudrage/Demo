To create reader and writer objects in the csv module in Python, you pass a file object as an argument to csv.reader() and csv.writer(). These file objects should be opened in the appropriate mode for reading or writing CSV data.

# Q3

File objects for reader and writer objects need to be opened in the following modes:
- For a reader object: Open the file in text mode ('r').
- For a writer object: Open the file in text mode ('w' or 'a' for append).

The writerow() method takes a list argument and writes it to a CSV file. Each element of the list corresponds to a column in the CSV file, and writerow() writes a new row in the CSV file using the elements of the list as values.

# Q5

- delimiter: The delimiter keyword argument specifies the character used to separate fields (columns) in the CSV file. By default, it is set to a comma (,), but you can specify a different character like a tab (\t) or a semicolon (;) if your CSV file uses a different delimiter.
- line terminator: The line terminator keyword argument specifies the character or characters used to terminate each line in the CSV file. By default, it uses the platform-specific line terminator (e.g., '\r\n' on Windows, '\n' on Unix-like systems), but you can specify a custom line terminator if needed.
